```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['figure.figsize'] = (15.0, 10.0)
```

```python
loss_data = pd.read_csv('./data/loss.csv', names=['loss'])
```

```python
loss_data.plot(figsize = (15.0, 10.0))
plt.xlim(-1, 210)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid()
plt.show();
```

```python
loss_data['residual'] = loss_data['loss'] - loss_data['loss'].min()
```

```python
loss_data['residual'].plot(figsize = (15.0, 10.0))
plt.xlim(-1, 30)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid()
plt.show();
```

Как видно из графика, после 15 эпохи сеть начинает переобучаться
