```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['figure.figsize'] = (15.0, 10.0)
```

```python
path = './data/elu_max/'
```

```python
df = pd.DataFrame()
```

```python
for seed in range(10):
    df['seed_'+ str(seed)] = pd.read_csv(path + 'example' + str(seed) + '.csv', 
                                         names=['seed_'+str(seed)])['seed_'+ str(seed)]
df.plot()
plt.ylim(0.95, 0.995)
plt.xlim(-10, 210)
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title("ELU + MaxPool + Adam optimizer")
plt.grid()
plt.show();
```

```python
plt.plot(np.linspace(0.992, 0.992, 200))
plt.plot(df.to_numpy().mean(axis=1))
plt.ylim(0.95, 0.995)
plt.xlim(-10, 210)
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title("Average Accuracy")
plt.legend(['Target = 0.992', 'avg valid acc'])
plt.grid()
plt.show();
```

```python
plt.plot(np.linspace(0, 200, 200), df.to_numpy().var(axis=1))
plt.xlabel('Epoch')
plt.ylabel('Variance')
plt.grid();
```
